**Orden de vista**

Para BFS y DFS se usara el orden URDL (Up, Right, Down, Left):

Adem谩s, cuando el destino se encuentra en otra planta, se prioriza el salto por ascensor (si se est谩 sobre uno) 

```python

#Orden de insercion en la frontera: Up Right Down Left
DESPLAZAMIENTOS_URDL = [(-1,0), (0,1), (1,0), (0,-1)]


def acciones_disponibles(nodo):
    if debug:
        print("LOCALIZACION:  ENTRANDO EN LA BUSQUEDA DE ACCIONES DISPONIBLES")
    pl, x, y = nodo.get_coordenadas()
    acciones_disponibles = []
    
    #Salto de planta si se encuentra en la casilla de ascensor, a帽adir la casilla en las plantas disponibles
    #por heuristica deberian ser mas atractivas las de la planta del destino
    #y para las no informadas a帽adirlas antes que las demas para que se prioricen 
    if plantas[pl][x][y] == 'E':
        
        for planta in range(PLANTAS):
            
            if planta != pl and plantas[planta][x][y] == 'E':
                
                acciones_disponibles.append(((planta, x, y), COST_ASC))
                
                
                    
                    
    #A帽adir una accion por cada casilla de movimiento disponible siempre que se encuentre en el mapa y no sea pared
    for disp_x,disp_y in DESPLAZAMIENTOS_URDL:
        
        nuevo_x = x + disp_x
        nuevo_y = y + disp_y
    
        
        
        #Se encuentra en el rango y no es una pared
        if 0 <= nuevo_x < FILAS and 0 <= nuevo_y < COLUMNAS:
            casilla,coste = tipo_casilla(pl,nuevo_x,nuevo_y)
            if casilla != "#":
                acciones_disponibles.append(((pl,nuevo_x,nuevo_y),coste))
    
    if debug:
                    print(f"Las acciones disponibles del nodo de las coordenadas {nodo.get_coordenadas()} son :{acciones_disponibles}")        
    
    return acciones_disponibles   
```

Heuristica manhattan:

Se ha elegido Manhattan como heur铆stica principal porque:
1. El movimiento permitido es 煤nicamente ortogonal (URDL),
2. Manhattan es consistente, admisible y muy apropiada para un mapa de este estilo,
3. Es ideal cuando existen m煤ltiples caminos equivalentes entre dos puntos.

El entorno multinivel, se ha pensado para contemplar el uso de ascensores:
si el robot y el destino est谩n en distinta planta, la heur铆stica considera el mejor caso posible entre todos los ascensores disponibles (m铆nimo te贸rico), sumando el coste de subir.

Esta heur铆stica no elige el ascensor, solo proporciona una cota inferior admisible.
El algoritmo A* es quien decide realmente cu谩l es el mejor ascensor usando el coste real acumulado, permitiendo asi el uso de multiples ascensores


```python
#Calculo de manhattan
def manhattan(ax, ay, bx, by):
    return abs(ax-bx) + abs(ay-by)

#heuristica de manhattan
def h_manhattan_estado(estado, destino):
    
    if debug:
        print("ENTRANDO EN EL CALCULO DE HEURISTICA MANHATTAN")
    
    pl, x, y = estado
    pld, xd, yd = destino
    
    #Si el objtivo esta en la misma planta se hace manhattan al destino unicamente
    if pl == pld:
        if debug:
            print("El objetivo esta en la misma planta")
        return manhattan(x, y, xd, yd)
    
    if not ASCENSORES:
        ("ERROR: No hay ascensores mapeados")
        return 999**9
    
    #Si el objetivo esta en una planta distinta se busca el mejor ascensor teorico(El menos lejano) para usarlo como heuristica de referencia
    #y se hace manhattan del estado al ascensor mas el del ascensor a el destino mas el coste de usar el ascensor 
    mejor = 999**9
    if debug:
        print(f"Estas en la planta {pl} y el objetivo en la planta {pld}")
    for (xa, ya) in ASCENSORES:
        h = manhattan(x, y, xa, ya) + COST_ASC + manhattan(xa, ya, xd, yd)
        if h < mejor:
            mejor = h
    if debug:
        print(f"La heuristica del nodo {estado} es de {mejor}")
    return mejor





```

Expansion:
Creacion de los nodos de todas las acciones disponibles 


```python

def expand(frontier,nodo,heuristica,calculo_fn):
  # por cada acci贸n disponible almacenada en el atributo actions
  for accion, coste in nodo.get_acciones_disponibles():
    
    pl,x,y = accion
    tipo, _ =tipo_casilla(pl,x,y)
    # instanciamos un nuevo nodo
    n = Nodo()
    n.set_estado(tipo)
    n.set_coordenadas(accion)
    n.set_padre(nodo)
    n.set_accion(nodo.get_accion()+[accion])
    n.set_acciones_disponibles(acciones_disponibles(n))
  
    n.set_coste(coste)
    n.set_coste_acumulado(nodo.get_coste_acumulado()+ n.get_coste())
    n.set_heuristica(heuristica(accion,DESTINO))
    n.set_fn(calculo_fn(n))
    n.set_profundidad(nodo.get_profundidad() + 1)
    #  lo a帽adimos a la lista
    frontier.append(n)
   
    print(n)

    
  #Devolvemos lista
  return frontier
```
